```python
#
# 2 qubit circuit with ibmqx5
#

import sys, copy
import numpy as np
import pickle

from IBMQuantumExperience import IBMQuantumExperience
from qiskit import QuantumProgram, QuantumCircuit
import qiskit.tools.qcvv.tomography as tomo

sys.path.append("/Users/wadejong/Desktop/Quantum-Computing/IBM/qiskit-sdk-py") # go to parent dir
import Qconfig
qx_config = {"APItoken": Qconfig.APItoken, "url": Qconfig.config['url']}
    
api = IBMQuantumExperience(token=qx_config['APItoken'], config={'url': qx_config['url']})

qp = QuantumProgram()
qp.set_api(Qconfig.APItoken, Qconfig.config['url'])

qr = qp.create_quantum_register('qr', 2)
cr = qp.create_classical_register('cr', 2)

def assembled(base,action):
    circuit = copy.deepcopy(base)
    if (action == 1):
        circuit.measure(qr[0],cr[0])
    elif (action == 2):
        circuit.measure(qr[1],cr[1])
    elif (action == 3):
        circuit.ry(-np.pi/2, qr[0])
        circuit.ry(-np.pi/2, qr[1])
        circuit.measure(qr,cr)
    elif (action == 4):
        circuit.rx(np.pi/2, qr[0])
        circuit.rx(np.pi/2, qr[1])
        circuit.measure(qr,cr)
    return circuit
    
for Theta in np.linspace(-np.pi,np.pi,101):
    base_qc = QuantumCircuit(qr, cr)
    base_qc.x(qr[0])
    base_qc.ry(Theta, qr[1])
    base_qc.cx(qr[1], qr[0])

    qp.add_circuit(str(Theta)+'_Z0',assembled(base_qc,1))
    qp.add_circuit(str(Theta)+'_Z1',assembled(base_qc,2))
    qp.add_circuit(str(Theta)+'_X0X1',assembled(base_qc,3))
    qp.add_circuit(str(Theta)+'_Y0Y1',assembled(base_qc,4))

# The IBM backend doesn't like me to do all at once (doing batches of 50 circuits)

def batch(iterable, n = 1):
    current_batch = []
    for item in iterable:
        current_batch.append(item)
        if len(current_batch) == n:
            yield current_batch
            current_batch = []
    if current_batch:
        yield current_batch

all_circuits = qp.get_circuit_names()
all_real_data = []

batch_num = 1

for x in batch(all_circuits, 50):
    real_data_result = qp.execute(x, backend='ibmqx5', max_credits=5, shots=8192, wait=20, timeout=1800)
    all_real_data.append(real_data_result)
    for circuit in real_data_result.get_names():
        print('Name', circuit, real_data_result.get_counts(circuit))
    results_file = open('results_ibmqx5_batch_'+str(batch_num)+'.pkl', 'wb')
    pickle.dump(all_real_data, results_file)
    results_file.close()
    batch_num += 1
```

```python
# Results analysis with confusion matrix, confusion matrices build from one set of 8192 shots
#
# Using two qubit confusion matrix for all terms
#
# Linear solve of Ax=b with constrains of x values between 0 and 1
#

import numpy as np
import pickle
from scipy.optimize import lsq_linear
from IBMQuantumExperience import IBMQuantumExperience
from qiskit import QuantumProgram, QuantumCircuit, Result
import qiskit.tools.qcvv.tomography as tomo

result_data = pickle.load( open('results_ibmqx5_all.pkl', 'rb') )

angles, energies = [], []
Z0, Z1, X0X1, Y0Y1 = [], [], [], []
bounds, x0 = [[0.,1.]]*4, [0.25]*4
all_data = {}

def fn(x, A, b):
    return (np.linalg.norm(A.dot(x) - b))

for result in result_data:
    for circuit in result.get_names():
        angle , expectation_value = circuit.split("_")
        numbers = [0, 0, 0, 0]
        b = np.zeros([4])
        for key, value in result.get_counts(circuit).items():
            b[int(key,2)] = value 
        xx = minimize(fn, x0, args=(confusion_2qb, b), method='SLSQP',bounds=bounds).x
        if expectation_value == 'Z0':
            angles.append(float(angle))
            answer = xx[0]+xx[2]-xx[1]-xx[3]
            Z0.append(answer)
        elif expectation_value == 'Z1': 
            answer = xx[0]+xx[1]-xx[2]-xx[3]
            Z1.append(answer)
        elif expectation_value == 'X0X1':
            answer = xx[0]+xx[3]-xx[1]-xx[2]
            X0X1.append(answer)
        else: # Y0Y1
            answer = xx[0]+xx[3]-xx[1]-xx[2]
            Y0Y1.append(answer)
        if angle in all_data.keys():
            all_data[angle].append([expectation_value, answer])
        else:
            all_data[angle] = [[expectation_value, answer]]

for val_z0, val_z1, val_x0x1, val_y0y1 in zip(Z0,Z1,X0X1,Y0Y1):
    energies.append(5.90670945+0.21829054999999986*val_z0-6.125*val_z1-2.1433035249352805*(val_x0x1+val_y0y1))

import matplotlib.pyplot as plt
from scipy import interpolate
from scipy.optimize import fmin
%matplotlib inline

plt.figure(1)
plt.subplot(211)
plt.xlabel("Theta")
plt.ylabel("<O>")
plt.plot(angles, Z0, label='Z0')
plt.plot(angles, Z1, label='Z1')
plt.plot(angles, X0X1, label='X0X1')
plt.plot(angles, Y0Y1, label='Y0Y1')
plt.legend()

plt.subplot(212)
plt.xlabel("Theta")
plt.ylabel("Energy")
plt.plot(angles, energies)
plt.axis([-np.pi, np.pi, -2, 14])
plt.yticks(np.arange(-2, 16, 2))
plt.show()

print('Z0 min',min(Z0),'and max',max(Z0))
print('Z1 min',min(Z1),'and max',max(Z1))
print('X0X1 min',min(X0X1),'and max',max(X0X1))
print('Y0Y1 min',min(Y0Y1),'and max',max(Y0Y1))
print('Calculated minimum energy = ', min(energies))

def fit_min(u):
    return (interpolate.splev(u, fit))
    
fit = interpolate.splrep(angles, energies)
xmin = fmin(fit_min,0.6, maxiter=10000, maxfun=10000)
print('Xmin',xmin,'Value',fit_min(xmin))
```
